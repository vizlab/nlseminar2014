### 18.

```python
from nltk.corpus import gutenberg
```

```python
text = nltk.Text(gutenberg.words())
```

```python
text[50:]
```

```python
nltk.regexp_tokenize(text, r'^wh')
```

```python
p = re.compile(r'^wh\w*')
```

```python
fd = nltk.FreqDist(word for word in text if p.match(word))
```

```python
fd.items()
```

### 19.

```python
f = open('freqdist.txt', 'w')
for k, v in fd.items():
    f.write('{0} {1}\n'.format(k, v))
f.close()
```

```python
freq = [[word, int(freq)] for word, freq in [row.split() for row in open('freqdist.txt').readlines()]]
```

### 21.

```python
def unknown(url):
    text = html_text_from_url(url)
    words = set(word.lower() for word in nltk.regexp_tokenize(text, r'[a-zA-Z]+'))
    return words - set(nltk.corpus.words.words())
```
